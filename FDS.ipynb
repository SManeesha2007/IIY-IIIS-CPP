```python
import pandas as pd
data={'ID':[101,102,103,104,105],'Name':['Alice','Bob','Charlie','David','Eve'],'Age':[25,None,29,33,None],'salary':[55000,48000,None,62000,None]}
df=pd.DataFrame(data)
print(df)
#remove rows with missing values
print("removing rows with missing values.................")
df_cleaned=df.dropna()
print(df_cleaned)
```

```python
for column in df.columns:
  if df[column].dtype=='float64' or df[column].dtype=='int64':
    df[column].fillna(df[column].mean(),inplace=True)
print(df)
```

```python
for column in df.columns:
  if df[column].isnull().any():
    df[f'(column)_Flag']=df[column].isna().apply(lambda x:'Missing' if x else 'Not Missing')
df['Missing_Flag']=df.isnull().any(axis=1).apply(lambda x:'Missing' if x else 'Not Missing')
print(df)
```

```python
import pandas as pd
data={'ID':[101,102,103,104,102,106,101],'Name':['Alice','Bob','Charlie','David','Bob','Alice','Alicia'],'Age':[25,29,29,33,29,27,25],'salary':[55000,48000,55000,62000,48000,59000,55000],'City':['NY','LA','NY','SF','LA','LA','NY']}
df=pd.DataFrame(data)
print("Original Dataset")
print(df,'\n')

df_cleaned=df.drop_duplicates()
print("Duplicates Removed -Exact Match")
print(df_cleaned,'\n')

#partial Match -based on specific columns only
df_cleaned_id=df.drop_duplicates(subset=['ID'])
print("Duplicates Removed -partial Match based on ID")
print(df_cleaned_id,'\n')

df_cleaned_name=df.drop_duplicates(subset=['Name'])
print("Duplicates Removed -partial Match based on Name")
print(df_cleaned_name,'\n')

df_cleaned_specific=df.drop_duplicates(subset=['ID','Name'])
print("Duplicates Removed -partial Match based on ID,Name")
print(df_cleaned_specific,'\n')
```
